# **Experiment 3:Implement the back propagation algorithm and train a network that computes the parity of 5 input bits**

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size

        self.weights_ih = np.random.randn(self.input_size, self.hidden_size)
        self.bias_h = np.zeros((1, self.hidden_size))
        self.weights_ho = np.random.randn(self.hidden_size, self.output_size)
        self.bias_o = np.zeros((1, self.output_size))

    def forward(self, x):
        self.hidden = sigmoid(np.dot(x, self.weights_ih) + self.bias_h)
        self.output = sigmoid(np.dot(self.hidden, self.weights_ho) + self.bias_o)
        return self.output

    def backward(self, x, y, output, learning_rate):
        error = y - output
        d_output = error * sigmoid_derivative(output)

        error_hidden = np.dot(d_output, self.weights_ho.T)
        d_hidden = error_hidden * sigmoid_derivative(self.hidden)

        self.weights_ho += learning_rate * np.dot(self.hidden.T, d_output)
        self.bias_o += learning_rate * np.sum(d_output, axis=0, keepdims=True)
        self.weights_ih += learning_rate * np.dot(x.T, d_hidden)
        self.bias_h += learning_rate * np.sum(d_hidden, axis=0, keepdims=True)

    def train(self, x, y, epochs, learning_rate):
        for _ in range(epochs):
            output = self.forward(x)
            self.backward(x, y, output, learning_rate)

    def predict(self, x):
        return np.round(self.forward(x))

def generate_parity_data(num_samples):
    x = np.random.randint(0, 2, (num_samples, 5))
    y = np.sum(x, axis=1) % 2
    return x, y.reshape(-1, 1)

if __name__ == "__main__":
    x_train, y_train = generate_parity_data(100)

    nn = NeuralNetwork(input_size=5, hidden_size=8, output_size=1)
    nn.train(x_train, y_train, epochs=100000, learning_rate=0.01)

    x_test, y_test = generate_parity_data(20)
    predictions = nn.predict(x_test)

    for i in range(len(x_test)):
      input_bits = x_test[i]
      expected_parity = y_test[i][0]
      predicted_parity = predictions[i][0]
      print(f'Input:{input_bits}, Expected:{expected_parity}, Prediced:{predicted_parity}')

    accuracy = np.mean(predictions==y_test)
    print(f'\nAccuracy:{accuracy*100:.2f}%')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skopt import gp_minimize
from skopt.space import Real
from skopt.utils import use_named_args
from scipy.signal import find_peaks
from skopt.plots import plot_convergence
```

```python
a = pd.read_csv('/content/corona virus.csv')
x = a.iloc[:,4:7].values.reshape(-1,1)
y = a.iloc[:,-1].values.reshape(-1,1)
```

```python
search_space=[Real(-1,1, name="x"), Real(-1,1, name="y")]
```

```python
def objective(X):
  return -five_peaks(X)
```

```python
def five_peaks(X):
  x,y=X
  return -(x+y)
```

```python
result = gp_minimize(func=objective, dimensions=search_space, n_calls=50, n_random_starts=10, random_state=42)
```

```python
print(f'Best Parameter found x: {-result.x[0]:.4f}')
print(f'Maximum value: {-result.fun:.4f}')
```

```python
function_values = -np.array(result.func_vals)
iterations = np.arange(1, len(function_values)+1)

peaks,_ = find_peaks(function_values)
minima,_ = find_peaks(-function_values)

fig,axs = plt.subplots(1,2, figsize=(12,6))

axs[0].plot(iterations, function_values, color="green", label="convergence plot")
axs[0].scatter(iterations[peaks], function_values[peaks], color="blue", label="Peaks")
axs[0].scatter(iterations[minima], function_values[minima], color="yellow", label="minima")
axs[0].set_title("Minima and Maxima plot")

plot_convergence(result, ax=axs[1])
axs[1].set_title("Convergence plot")
plt.show()
```

# **Expermint 5:Generate a Feed forward neural network on a dummy data.**

```python
from sklearn.datasets import make_classification
x, y = make_classification(n_samples=1000, n_features=20, n_classes=2, n_informative=10, n_redundant=5, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 42)
```

```python
from tensorflow.keras import layers, models
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(x_train.shape[1],)))
model.add(layers.Dense(32, activation="relu"))
model.add(layers.Dense(32, activation="relu"))
model.add(layers.Dense(32, activation="relu"))
model.add(layers.Dense(1, activation="sigmoid"))
```

```python
model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics=["accuracy"])
```

```python
result = model.fit(x_train, y_train, epochs=50, batch_size=32, validation_data=(x_test,y_test))
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt
fig, axis = plt.subplots(1,2,figsize=(12,6))
axis[0].plot(result.history["accuracy"])
axis[0].plot(result.history["val_accuracy"])
axis[0].set_title("Accuracy")
axis[0].set_xlabel("Epochs")
axis[0].set_ylabel("Accuracy")
axis[1].plot(result.history["loss"])
axis[1].set_title("Loss")
axis[1].set_xlabel("Epochs")
axis[1].set_ylabel("Loss")
plt.show()
```
